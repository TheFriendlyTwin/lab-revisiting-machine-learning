1. Complete the following steps on the categorical columns in the dataset:

- Check for null values in all the columns

- Exclude the following variables by looking at the definitions. Create a new empty list called drop_list. We will append this list and then drop all the columns in this list later:

    - OSOURCE - symbol definitions not provided, too many categories
    - ZIP CODE - we are including state already

```python
# Check for null values in each column
null_counts = df.isnull().sum()

# Display the count of null values per column
print(null_counts)
```

```python
#Initializing an empty list to append the columns to drop
drop_list = []
```

```python
# Append both column names to drop at once
drop_list.extend(['OSOURCE', 'ZIP'])

# Now, columns_to_drop contains the names of the columns you want to drop later
print(drop_list)
```

```python
#Dropping the previous columns from our database
df_copy = df.copy()

df_copy = df_copy.drop(columns=drop_list)

df_copy
```

2. Identify columns that over 85% missing values

```python
# Calculate the percentage of missing values for each column
missing_percentage = df_copy.isnull().sum() / len(df_copy) * 100

# Identify columns where the percentage of missing values is greater than 85%
missing_values_columns = missing_percentage[missing_percentage > 85].index.tolist()

# Print the list of column names with more than 85% missing values
print(missing_values_columns)
```

3. Remove those columns from the dataframe

```python
# Removing the missing values columns from our dataframe
df_copy = df_copy.drop(columns=missing_values_columns)

df_copy
```

4. Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest

  - Note that there are a few null values in the column. We will first replace those null values using the code below:

  print(categorical['GENDER'].value_counts())
  
  categorical['GENDER'] = categorical['GENDER'].fillna('F')

```python
print(df_copy['GENDER'].value_counts())
df_copy['GENDER'] = df_copy['GENDER'].fillna('F')
```

```python
# Replace values in 'gender' column where the value is not 'F' or 'M' with 'O'
df_copy.loc[~df['GENDER'].isin(['F', 'M']), 'GENDER'] = 'O'

print(df_copy['GENDER'].value_counts())
```
